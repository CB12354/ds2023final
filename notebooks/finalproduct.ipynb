# Final Product: *Where in the Wordle?*
![Wordle Poster](../assets/poster_notebook_files/wordleposter.png)  

Image file: [wordleposter.png](../assets/poster_notebook_files/wordleposter.png)

*Where in the Wordle?* is a poster exploring the etymology of historical Wordle puzzles through maps of very different forms. Both where they come from and which languages are related are covered, the former by the *geographical* map and the latter by a *heat*map. It's meant to sell just how varied English word origins are, even from this small subset of 5-letter words. People complain about how complex and inconsistent English is, but rarely question why. It's because of this very statistic - that there are 60 languages in a trench coat trying to pass off as one language and make their different grammar, syntax, and plurality rules mesh together in one extremely inconvenient-to-remember umbrella.

I tried to make this poster be as inspired by its data as possible, both in form and function. I used the same color scheme and fonts as Wordle and kept the presentation (excluding the Mercator map) as clean and minimalist as possible, like how the game is presented on the New York Times with basic geometric shapes. I also did my best to ensure brevity in my descriptions while ensuring readers could still understand it, since Wordle is a fast-paced game that anyone can pick up and understand.
